```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
```

```python
#Loading the Datas
train_dt = pd.read_csv('train.csv')
test_dt=pd.read_csv('test.csv')
```

```python
test_dt.dtypes
numeric_columns = train_dt.select_dtypes(include=['float64', 'int64']).columns

# Fill missing values in training set
train_dt[numeric_columns] = train_dt[numeric_columns].fillna(train_dt[numeric_columns].mean())

```

```python
square_footage_columns = ['GrLivArea','TotalBsmtSF','GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch',
                           '3SsnPorch', 'ScreenPorch', 'PoolArea']
```

```python
train_dt['TotalSquareFootage'] = train_dt[square_footage_columns].sum(axis=1)
test_dt['TotalSquareFootage'] = test_dt[square_footage_columns].sum(axis=1)
```

```python
X_train= train_dt[['TotalSquareFootage','BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr' ]]
```

```python
y_train=train_dt['SalePrice']
```

```python
X_test=test_dt[['TotalSquareFootage','BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr' ]]
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# Impute missing values in the test set
imputer = SimpleImputer(strategy='mean')
X_test_scaled_imputed = imputer.fit_transform(X_test_scaled)
```

```python
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train_scaled, y_train, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train_split, y_train_split)
```

```python
y_val_pred=model.predict(X_val)
```

```python
#CALCULATING MEAN SQUARED ERROR
mse= mean_squared_error(y_val, y_val_pred)
print(f'Mean Squared Error : {mse}')
```

```python
y_test_predicted_scaled = model.predict(X_test_scaled_imputed)
```

```python
print(f'Predicted House prices for the data : {y_test_predicted_scaled}')
```

```python
avg_pred= y_test_predicted_scaled.mean()
```

```python
print(f'Average Predicted Price: {avg_pred} ')
```
